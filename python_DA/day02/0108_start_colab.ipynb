
1. 유료 사용자(is_paid == True)의 나이만 출력해 보세요.

2. 나이가 30세 이상인 사용자의 수를 출력해 보세요.

```python
cnt = 0
for user in users:
  if user['age'] >= 30:
    cnt += 1
cnt
```

3. 전체 사용자의 평균 나이를 계산해 보세요.

```python
sum_age = 0
for user in users:
  sum_age += user['age']

avg_age = sum_age/len(users)
avg_age
```

4. 유료 사용자만의 평균 나이를 계산해 보세요.

```python
paid_age = 0
paid_cnt = 0
for user in users:
  if user['is_paid'] == True:
    paid_age += user['age']
    paid_cnt += 1

avg_age2 = paid_age/paid_cnt
print(round(avg_age2,1))
```

5. 유료 vs 무료 사용자 평균 나이 비교

```python
paid_age = 0
paid_cnt = 0
free_age = 0
free_cnt = 0

for user in users:
  if user['is_paid'] == True:
    paid_age += user['age']
    paid_cnt += 1
  else:
    free_age += user['age']
    free_cnt += 1

paid_avg = paid_age / paid_cnt
free_avg = free_age / free_cnt

print("paid_avg", paid_avg)
print("free_avg", free_avg)
```
