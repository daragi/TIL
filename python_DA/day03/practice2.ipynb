[4] 함수로 값 변환하기

- 'gender' 컬럼의 값을 'male'은 'M', 'female'은 'F'로 변경하고 titanic에 적용하세요.

```python
titan.head()
```

```python
titan['gender'] = titan['gender'].map({'male': 'M', 'female' : 'F'})
titan
```

문자열 전처리 없이도 “값 표준화”는 할 수 있습니다. embarked 컬럼 값을 다음 규칙으로 치환해 embarked_name 컬럼을 새로 만드세요.
- S → Southampton
- C → Cherbourg
- Q → Queenstown

```python
titan['embarked_name'] = titan['embarked'].map({'S': 'Southampton', 'C': 'Cherbourg', 'Q': 'Queenstown'})
titan
```
