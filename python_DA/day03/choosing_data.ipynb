### (2) 다중 조건

- 비트 연산자 사용

```python
con1 = df['Occupation'] == 'Artist'
con2 = df['Age'] >= 30
```

```python
df[con1 & con2]
```

```python
df[con1 | con2]
```

```python
df[con1 ^ con2]
```

```python
df.loc[con1 & con2, ['Rating','Age']]
```

```python
# loc 문법하에서, 행 필터링하는데 사용 가능
fr = df.loc[con1 & con2,['Occupation','Age','Rating']]
fr.head()
```

```python
df.loc[df['Occupation']=='Artist','Rating'].value_counts()
```

```python
df.loc[df['Occupation']=='Artist','Rating'].mean()
```

```python
df.loc[~(df['Occupation']=='Artist'),'Rating'].mean()
```
