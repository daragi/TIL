생존한 승객 중(survived)에서 각 등급(pclass)별 승객 수를 구하세요.


타이타닉호의 승객 중에서 형제자매/배우자 (sibsp 참고)가 없고, 20세 이하인 승객의 평균 요금을 계산하세요.
