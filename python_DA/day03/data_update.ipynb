## [4] 자료형 변경하기

```python
df = pd.DataFrame({'name': ['alex','jun','chelsea'],
                   'age': ['12','15','29'],
                   'is_license': ['True', 'False', 'False'],
                   'playtime': ['12.54', '11','10']})
df
```

`astype()`은 컬럼 단위로 사용 가능
- 변환 불가한 값이 있으면 에러가 발생

```python
# 단일 컬럼 변경
df['age'] = df['age'].astype(int)
# inplace = True 는 불가
df.dtypes
```

```python
bool('False')
# False 문자열은 비어있는 형태가 아니기에 boolean형 변환 시 True 반환
```
