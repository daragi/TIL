```python
# 최빈값
subjects = ["Python", "Java", "Python", "C++", "Python", 
            "Java", "C++", "Python", "Data Science", 
            "Data Science", "Python"]
sub_cnt = {}
for subject in subjects:
    if subject not in sub_cnt:
        sub_cnt[subject] = 1
    else:
        sub_cnt[subject] += 1
sub_cnt
```

```python
# 최빈값 찾기
high_sub = ""
high_cnt = 0

for sub, cnt in sub_cnt.items():
    if cnt > high_cnt:
        high_cnt = cnt
        high_sub = sub

print(f'최빈과목은 {high_sub}이고, 횟수는 {high_cnt}회이다.')
```

```python
# 사분위수 구하기
scores = [55, 60, 65, 70, 75, 80, 85, 90, 95, 100, 37, 40, 20, 10, 90]

scores.sort()
len(scores)
```

```python
# Q2(중앙값 구하기)
scores[len(scores)//2]
# Q1 (25%)
Q1 = scores[3]
# Q3 (75%)
Q3 = scores[-4]

IQR = Q3 - Q1
IQR
```

```python
# 분산과 표준편차
mean = sum(scores)/len(scores)
var = 0

for score in scores:
    var += (score - mean)**2
var = var/len(scores)
print(round(var,2))

std = var ** 0.5
print(round(std,2))
```
